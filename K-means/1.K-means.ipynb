```python
import numpy as np
import pandas as pd
from matplotlib import pylab as plt
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from scipy import io as sio

%matplotlib inline
```

```python
mat = sio.loadmat('./data/ex7data2.mat')
mat.keys()
```

```python
df = pd.DataFrame(mat.get('X'), columns=['X1', 'X2'])
df.head()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df['X1'], df['X2'])
```

```python
kmeans = KMeans(n_clusters=3, n_init=500, max_iter=2000)
```

```python
kmeans.fit(df.values)
kmeans.inertia_
```

```python
classes = kmeans.predict(df.values)
classes.shape
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(df['X1'], df['X2'], c=classes)
plt.scatter(centers[:, 0], centers[:, 1], marker='X', linewidths=10)
plt.xlim((-1, 9))
plt.ylim((-1, 9))
```
