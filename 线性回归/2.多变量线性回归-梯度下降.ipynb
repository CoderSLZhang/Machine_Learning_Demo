```python
import numpy as np
import pandas as pd
from matplotlib import pylab as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import SGDRegressor
from sklearn import metrics
```

```python
PATH =  'ex1data2.txt'
df = pd.read_csv(PATH, header=None, names=['Size', 'Bedrooms', 'Price'])

df.head()
```

```python
df.describe()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:, [0, 1]].values, df.iloc[:, -1].values, test_size=0.2)
```

```python
print('X_train:', X_train.shape)
print('X_test:', X_test.shape)
print('Y_train:', Y_train.shape)
print('Y_test:', Y_test.shape)
```

```python
X_scaler = StandardScaler()
Y_scaler = StandardScaler()

X_train = X_scaler.fit_transform(X_train.reshape(-1, 2))
X_test = X_scaler.transform(X_test.reshape(-1, 2))
Y_train = Y_scaler.fit_transform(Y_train.reshape(-1, 1))
Y_test = Y_scaler.transform(Y_test.reshape(-1, 1))
```

```python
regressor = SGDRegressor(max_iter=1000, verbose=True)
```

```python
regressor.fit(X_train, Y_train)
```

```python
regressor.score(X_train, Y_train)
```

```python
scores = cross_val_score(regressor, X_train, Y_train, cv=5)
```

```python
scores
```

```python
regressor.score(X_test, Y_test)
```

```python
predicts_test = regressor.predict(X_test)
```

```python
metrics.mean_squared_error(Y_test, predicts_test)
```
